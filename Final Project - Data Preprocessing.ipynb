# NYS Residential Homes Energy Efficiency Projects

> The Green Jobs - Green New York Program provides New Yorkers with access to energy assesments, installation services, low interest financing, and pathway to training for various green-collar careers

## Columns Explanation:

- Reporting Period : The time period covered by the data set
- Home Performance Project ID : Unique identifier for project	
- Home Performance Site ID	: Unique site identifier for location of the project	
- Project County : Name of county for project location
- Project City : Name of city for project location
- Project Zip : Name of zip code for project location
- Gas Utility : Name of gas utility for project location. If blank, then utility was not reported, or project location is not served by a gas utility
- Electric Utility	: Name of electric utility for project location
- Project Completion Date : Date final project completion paperwork was reviewed and approved by Program.
- Customer Type	: Indicates if the homeowner received market rate incentives or the assisted subsidy for households with income up to 80% of the State or median county income, whichever is higher
- Low-Rise or Home Performance Indicator : Indicates whether project was part of a low-rise community initiative or a single-family home performance project. Low-rise projects are not eligible for residential loan products, or free or reduced cost audits. Each unit served on a low-rise project is treated as an individual project
- Total Project Cost : Cost of project in US dollars (USD)
- Total Incentives : Total Incentives Numeric Amount of homeowner incentive (USD). Zero dollars represent either projects receiving only financing, or Market Rate projects that did not receive an incentive, but still participated in the program to take advantage of working with participating contractors and Quality Assurance benefits
- Type of Program Financing	: Indicates type of Program financing a project received, including current financing options of GJGNY On-Bill Recovery Loan, GJGNY Smart Energy Loan, and legacy financing options (Energy Smart Loan Fund (ESMART) and Home Performance with ENERGY STAR Financing (ESTAR)). If blank, then project was not financed through the GJGNY Smart Energy Loan Program, GJGNY On-Bill Recovery Loan or other legacy Program financing option.
- Amount Financed Through Program : Project loan amount (USD). If blank, then project was not financed through the GJGNY Smart Energy Loan Program, GJGNY On-Bill Recovery Loan or other legacy Program financing option.
- Pre-Retrofit Home Heating Fuel Type	: Indicates the pre-retrofit primary heating fuel type
- Year Home Built	: Home construction date. 
- Size of Home	: Square footage of home. 
- Volume of Home : Approximate volume of conditioned space of the home. 
- Number of Units : Number of units served by the Program .
- Measure Type	: Measure classification describing primary project improvement.
- Estimated Annual kWh Savings	: Annual post-retrofit modeled electric savings estimate in kWh. Negative numbers represent projects with post-retrofit increase in electric consumption, typically from fuel conversions or ancillary savings
- Estimated Annual MMBtu Savings : Annual post-retrofit modeled MMBtu savings based on primary fuel type. Negative numbers represent projects with post-retrofit increase in MMBtu consumption, typically from fuel conversions or ancillary savings
- First Year Energy Savings Estimate : Estimated modeled post-retrofit first year dollar savings (USD). Negative numbers represent projects with estimated post-retrofit first year dollar expenses, typically occurring when non-energy work was completed such as health and safety improvements, or when work was done in conjunction with another, net positive energy savings project
- Homeowner Received Green Jobs-Green NY Free/Reduced Cost Audit (Y/N)	: Indicator if the homeowner received a Free/Reduced Cost Green Jobs Green-NY audit prior to retrofit work
- Location	: Municipal location based on supplied address location.

## Handling Missing Data

**Drop the columns that will not be use**

```python
df.drop(['Location', 'New York Zip Codes', 'Counties',
       'NYS Municipal Boundaries','Home Performance Project ID',
       'Home Performance Site ID','Reporting Period','Project Zip','Year Home Built'],axis=1,inplace=True)
```

**Fill NaN Values in Project County Columns**

```python
for i in range(len(df)):
    if str(df.loc[i,'Project County'])=='nan':
        if (df.loc[i,'Project City']) == 'Brewerton' :
            df.loc[i,'Project County'] = 'Onondaga'
        elif (df.loc[i,'Project City']) == 'Tonawanda' :
            df.loc[i,'Project County'] = 'Erie'
```

**Fill NaN Values in Amount Financed Through Program columns**

```python
for i in range(len(df)):
    if str(df.loc[i,'Amount Financed Through Program'])=='nan':
        df.loc[i,'Amount Financed Through Program'] = 0.0
```

**Fill NaN Values in Gas Utility columns**

```python
df['Gas Utility'].fillna('No Gas Provider',inplace=True)
```

**Fill NaN Values in Type of Program Financing columns**

```python
df['Type of Program Financing'].fillna('Not Financed Through The Legacy Program Financing Option',inplace=True)
```

**Fill NaN Values in Volume of Home columns and Size of Home colums**

```python
meanVolume = df.groupby('Size of Home').mean()['Volume of Home']
def fillVolume(x):
    forfill = []
    for i in df['Size of Home']:
        if x['Volume of Home'] > 0:
            return x['Volume of Home']
        else:
            if x['Size of Home'] > 0:
                return meanVolume.loc[x['Size of Home']]
            else:
                return np.nan
df['Volume of Home'] = df.apply(fillVolume,axis=1)
```

```python
meanSize = df.groupby('Volume of Home').mean()['Size of Home']
def fillSize(x):
    forfill = []
    for i in df['Volume of Home']:
        if x['Size of Home'] > 0:
            return x['Size of Home']
        else:
            if x['Volume of Home'] > 0:
                return meanSize.loc[x['Volume of Home']]
            else:
                return np.nan
df['Size of Home'] = df.apply(fillSize,axis=1)
```

> For Volume of Home column and Size of Home column, after the NaN Values filled using mean from groupby method, those columns still have some NaN Values. To handle the issue in those columns, the rest of missing data will fill by mean from each columns value after dropping the outlier value

```python
# Function to check outlier
def check_outlier(data, col):
    q1 = data[col].quantile(0.25)
    q3 = data[col].quantile(0.75)
    iqr = q3-q1
    upper_limit = q3 + (1.5 * iqr)
    lower_limit = q1 - (1.5 * iqr)
    outlier = data[(data[col] < lower_limit) | (data[col] > upper_limit)].index
    return outlier
```

```python
# Apply the function in Volume of Home columns
index_to_drop = check_outlier(df,'Volume of Home')
Not_Outlier = df.drop(index_to_drop)
# Fill NaN Values using mean from not outlier data
df['Volume of Home'].fillna(Not_Outlier['Volume of Home'].mean(),inplace=True)

# Apply the function in Size of Home columns
index_to_drop = check_outlier(df,'Size of Home')
Not_Outlier = df.drop(index_to_drop)
# Fill NaN Values using mean from not outlier data
df['Size of Home'].fillna(Not_Outlier['Size of Home'].mean(),inplace=True)
```

**Fill NaN Values in Number of Units column**

```python
Vol_AvgUnit4 = df[df['Number of Units']==4.0]['Volume of Home'].mean()
Vol_AvgUnit3 = df[df['Number of Units']==3.0]['Volume of Home'].mean()
Vol_AvgUnit2 = df[df['Number of Units']==2.0]['Volume of Home'].mean()
Vol_AvgUnit1 = df[df['Number of Units']==1.0]['Volume of Home'].mean()
print (f'Avg Volume for 4 Unit: {Vol_AvgUnit4}\n')
print (f'Avg Volume for 3 Unit: {Vol_AvgUnit3}\n')
print (f'Avg Volume for 2 Unit: {Vol_AvgUnit2}\n')
print (f'Avg Volume for 1 Unit: {Vol_AvgUnit1}')
for i in range(len(df)):
    if str(df.loc[i,'Number of Units'])=='nan':
        if df.loc[i,'Volume of Home'] >= Vol_AvgUnit4:
            df.loc[i,'Number of Units'] = 4.0
        elif df.loc[i,'Volume of Home'] >= Vol_AvgUnit3 and  df.loc[i,'Volume of Home'] < Vol_AvgUnit4:
            df.loc[i,'Number of Units'] = 3.0
        elif df.loc[i,'Volume of Home'] >= Vol_AvgUnit2 and  df.loc[i,'Volume of Home'] < Vol_AvgUnit3:
            df.loc[i,'Number of Units'] = 2.0
        elif df.loc[i,'Volume of Home'] < Vol_AvgUnit2:
            df.loc[i,'Number of Units'] = 1.0
```

**Drop rows for NaN Values and change duplicate values in Pre-Retrofit Home Heating Fuel Type columns**

```python
df.dropna(subset=['Pre-Retrofit Home Heating Fuel Type'],inplace=True)
df.loc[df[df['Pre-Retrofit Home Heating Fuel Type']=='Natural gas'].index,'Pre-Retrofit Home Heating Fuel Type'] = 'Natural Gas'
df.loc[df[df['Pre-Retrofit Home Heating Fuel Type']=='Electric'].index,'Pre-Retrofit Home Heating Fuel Type'] = 'Electricity'
df.loc[df[df['Pre-Retrofit Home Heating Fuel Type']=='Anthracite Coal'].index,'Pre-Retrofit Home Heating Fuel Type'] = 'Coal'
df.loc[df[df['Pre-Retrofit Home Heating Fuel Type']=='Bituminous Coal'].index,'Pre-Retrofit Home Heating Fuel Type'] = 'Coal'
df.reset_index(drop=True,inplace=True)
```

**Change type of some columns to make the data more understandable**

## Drop Outlier

```python
# See Outlier distribution use boxplot
plt.figure(figsize=(10,5))
sns.boxplot(data=df[['Total Project Cost', 'Total Incentives',
       'Amount Financed Through Program', 'Size of Home','Volume of Home']],orient='v')
plt.xticks(rotation=45)
plt.title('Outlier Data in Columns',size=20)
plt.show()
```

**Drop Outlier for each numerical columns**

```python
# List columns to check the outlier
cek_outlier = ['Total Project Cost', 'Total Incentives',
       'Amount Financed Through Program','Size of Home',
       'Volume of Home']
total_outlier = []
# function to gather index of outlier for each columns in cek_oulier list
for item in cek_outlier:
    idx_outlier = check_outlier(df,item)
    for i in idx_outlier:
        if i not in total_outlier:
            total_outlier.append(i)
# drop outlier based from index in list total_outlier
df = df.drop(total_outlier)
df.reset_index(drop=True,inplace=True)
```

**Drop Multivariate Outlier use Mahalanobis Distance**

```python
# Import library that needed
from mahalanobis import Mahalanobis
from scipy.stats import chi2
# Choose columns that we want to check the outlier by make a new dataframe
df_outlier = df[['Total Project Cost', 'Total Incentives',
       'Amount Financed Through Program','Size of Home',
       'Volume of Home']]
ppf_score = chi2.ppf((1-0.05),df=4)
# Create new column filled by Mahalanobis distance for each rows
df_outlier['mahala_pack'] = Mahalanobis(np.array(df_outlier),len(df_outlier)).distances
# Function to create label for outlier / non outlier data
def nama_label(col):
    if col > ppf_score:
        return 'Outlier'
    else:
        return 'Non Outlier'
df_outlier['Label'] = df_outlier['mahala_pack'].apply(nama_label)
# Drop multivariate outlier
idx_outlier = df_outlier[df_outlier['Label']=='Outlier'].index
df.drop(index=idx_outlier,inplace=True)
df.reset_index(drop=True,inplace=True)
```
